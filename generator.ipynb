```python
import random
from datetime import datetime
import csv
from constants import *
```

```python
book_isbn = random.sample(range(ISBN_LOW, ISBN_HIGH + 1), NUM_BOOKS)

book_title = []
for i in range(NUM_BOOKS):
    if i < DISTINCT_BOOK_TITLES:
        book_title.append(f"book_title{i + 1:05d}")
    else:
        book_title.append(f"book_title{random.randint(1, DISTINCT_BOOK_TITLES):05d}")
book_title.sort()

num_printed = [random.randint(PRINT_MIN, PRINT_MAX) * 1000 for _ in range(NUM_BOOKS)]

num_pages = [random.randint(PAGE_MIN, PAGE_MAX) for _ in range(NUM_BOOKS)]

price = [round(random.uniform(1, PRICE_MAX), 1) for _ in range(NUM_BOOKS)]

edition_number = [random.randint(1, EDITION_MAX) for _ in range(NUM_BOOKS)]

print_series = [random.randint(1, PRINT_SERIES_MAX) for _ in range(NUM_BOOKS)]

category_id = [random.randint(1, NUM_CATEGORIES) for _ in range(NUM_BOOKS)]
category_name = [CATEGORIES[category_id[i] - 1] for i in range(NUM_BOOKS)]
```

```python
publisher_id = []
publisher_name = []
for i in range(NUM_BOOKS):
    temp_num = random.randint(PUBLIHSER_ID_MIN + 1, PUBLISHER_ID_MAX)
    publisher_id.append(temp_num)
    publisher_name.append(f"publisher_name{temp_num:04d}")
publisher_id.sort()
publisher_name.sort()

city = []
for i in range(NUM_BOOKS):
    if i > 0:
        if publisher_id[i] == publisher_id[i - 1]:
            city.append(city[i - 1])
        else:
            city.append(random.randint(1, NUM_CITIES))
    else:
        city.append(random.randint(1, NUM_CITIES))

manager_name = []
for i in range(NUM_BOOKS):
    if i > 0:
        if publisher_id[i] == publisher_id[i - 1]:
            manager_name.append(manager_name[i - 1])
        else:
            manager_name.append(f"manager_name{random.randint(1, DISTINCT_MANAGER_NAMES)}")
    else:
        manager_name.append(f"manager_name{random.randint(1, DISTINCT_MANAGER_NAMES)}")
        
publisher_revenue = []
for i in range(NUM_BOOKS):
    publisher_revenue.append(PUBLISHER_SHARE * price[i] * num_printed[i])
```

```python
author_id = []
author_name = []
for i in range(NUM_BOOKS):
    temp_num = random.randint(AUTHOR_ID_MIN + 1, AUTHOR_ID_MAX)
    author_id.append(temp_num)
    author_name.append(f"author_name{temp_num:04d}")
author_id.sort()
author_name.sort()

author_age = []
for i in range(NUM_BOOKS):
    if i > 0:
        if author_id[i] == author_id[i - 1]:
            author_age.append(author_age[i - 1])
        else:
            author_age.append(random.randint(AGE_MIN, AGE_MAX))
    else:
        author_age.append(random.randint(AGE_MIN, AGE_MAX))

author_revenue = []
for i in range(NUM_BOOKS):
    author_revenue.append(AUTHOR_SHARE * price[i] * num_printed[i])
```

```python
publish_date = []
for i in range(NUM_BOOKS):
    random_year = random.randint(2000, 2023)
    random_month = random.randint(1, 12)
    random_day = random.randint(1, 28)
    publish_date.append(datetime(year=random_year, month=random_month, day=random_day))
```

```python
attributes = ["isbn",
              "book_title",
              "num_printed",
              "num_pages",
              "price",
              "edition",
              "print_series",
              "category_id",
              "category_name",
              "publisher_id",
              "publisher_name",
              "city",
              "manager_name",
              "publisher_revenue",
              "author_id",
              "author_name",
              "author_age",
              "author_revenue",
              "publish_date"]
```

```python
with open("data.csv", "w") as f:
    f.write("")

with open("data.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(attributes)
    for i in range(NUM_BOOKS):
        row = [book_isbn[i],
               book_title[i],
               num_printed[i],
               num_pages[i],
               price[i],
               edition_number[i],
               print_series[i],
               category_id[i],
               category_name[i], 
               publisher_id[i],
               publisher_name[i],
               city[i],
               manager_name[i],
               publisher_revenue[i],
               author_id[i],
               author_name[i],
               author_age[i],
               author_revenue[i],
               publish_date[i]]
        writer.writerow(row)
```
